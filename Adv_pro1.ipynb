### Importing Neccessary Libraries

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

### Loading the dataset

```python
df = pd.read_json("in_CATALOGS.json")
df
```

## Data Preprocesses

### Taking all the catalogs(columns) discription and their ucds

```python
data = {
    'ucds':[],
    'descriptions':[]
}
for col in df.columns:
    _u = df[col].ucds
    _d = df[col].descriptions
    data['ucds'].extend(_u)
    data['descriptions'].extend(_d)
```

### Checking the target(ucds)  size (length)

```python
len(data['ucds'])
```

### Visualizing the content of ucds

```python
print(data['ucds'])
```

### Checking the databases (columns) discription size(length)

```python
len(data['descriptions'])
```

### Cleanning the data (take out "None" from discription) and adding the rest text discription and ucds in a new list

```python
new_ucds_clean_list = []
new_desc_clean_list = []
number_of = len(data['ucds'])

for i in range(number_of):
    _ucd = data['ucds'][i]
    _dsc = data['descriptions'][i]
    if _dsc is None:
        continue
    else:
        new_ucds_clean_list.append(_ucd)
        new_desc_clean_list.append(_dsc)
```

### Checking the size of the new description list

```python
len(new_desc_clean_list)
```

### Checking the size of the new ucds list

```python
len(new_ucds_clean_list)
```

### Converting the discription text document in to a matrix of token counts by using of CountVectorizer

```python
dec_count_vec = CountVectorizer(stop_words='english',min_df=2)
```

```python
decx_count_vec = dec_count_vec.fit_transform(new_desc_clean_list)
```

```python
decx_count_vec.shape
```

### Extracting the feature names from the vectorized text

```python
dec_names = dec_count_vec.get_feature_names()
dec_names
```

### Creating a dataframe 

```python
df = pd.DataFrame(decx_count_vec.toarray(), index = new_desc_clean_list, columns = dec_names)
```

```python
df.head()
```

### Assignning the value of x and Y

```python
x = decx_count_vec
y = new_ucds_clean_list
```

### Spliting the data in to trainning and testing 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, train_size=0.8)
```

### Checking the trainning and testing size

```python
x_train.shape
```

```python
x_test.shape
```

### Creating object of Support vector machine classification algorithem 

```python
model = SVC()
```

###  (trainning the model) fitting the trainning data to the classification model

```python
model.fit(x_train,y_train)
```

### (Evaluation) testing the model performance

```python
model.score(x_test,y_test)
```

# Tunning the parameters

```python
model = SVC(C=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

### SVM gamma = 1

```python
model = SVC(gamma=10)
```

### SVM kernel = linear

```python
model = SVC(kernel='linear')
```

```python
model.score(x_test,y_test)
```
